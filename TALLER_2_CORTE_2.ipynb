#### Regresión Logística

Hallando el Trio entre las columnas Age, Sex, SibSp, Pclass, Embarked que tenga el mayor Pseudo R-squ

```python
aux = set()
max = 0
for i in df.iloc[:,[2,5,6,11,12]]:
  aux2 = set()
  aux.add(i)
  for j in df.iloc[:,[2,5,6,11,12]]:
    aux2.add(j)
    for k in df.iloc[:,[2,5,6,11,12]]:
       if (i != j and j != k and k != i and not(k in aux2) and not(j in aux)):
        #Logit Regression
        x = df.loc[:,[i,j,k]]
        y = df.Survived
        logit_model = sm.Logit(y,x)
        result = logit_model.fit()
        print(result.summary())
        if max < result.prsquared:
          max = result.prsquared
          columnas = i + ' ' + j + ' ' + k
print('El mejor Pseudo R-squ es {} vs Survived: {}'.format(columnas,max))
```

RTA: El mejor Trio es Pclass, SibSp y Sex2 y se obtiene un Pseudo R-squ de 0.29389178324802556
