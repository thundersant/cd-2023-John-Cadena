```python
import numpy as np
import pandas as pd
import seaborn as sb
import os
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/fmoreno4/datasets/main/comprar_alquilar_con_cate.csv')
df
```

```python
df.isnull().sum()
```

```python
df.groupby(['trabajo']).count()
```

```python
#Transformación de datos
mapa1 = { 'Soltero': 1, 'Casado': 2, 'Separado': 3}
df['estado_civil'] = df['estado_civil'].map(mapa1)

mapa2 = { 'Tipo 0': 0, 'Tipo 1': 1, 'Tipo 2': 2, 'Tipo 3': 3, 'Tipo 4': 4, 'Tipo 5': 5, 'Tipo 6': 6, 'Tipo 7': 7,'Tipo 8': 8}
df['trabajo'] = df['trabajo'].map(mapa2)

df
```

```python
#Selección de variables
predictores = df[['ingresos','gasto_ppal','vehiculo','otros_gastos','ahorros','Valor_vivienda','estado_civil','hijos','trabajo']]
objetivo = df.comprara
```

```python
#Partición para validación cruzada
pred_train, pred_test, res_train, res_test = train_test_split(predictores, objetivo, test_size=0.2)
```

```python
#Modelo SVM
SVM_Model = SVC(gamma='auto', kernel='rbf')
SVM_Model.fit(pred_train, res_train)
```

```python
print (f'Precisión: {SVM_Model.score(pred_test, res_test):.4f}')
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/adiacla/bigdata/master/titanic.csv")
df
```

```python
df = df.drop('Cabin', axis = 1)
df
```

```python
df = df.dropna()
df
```

```python
df['Sex2'] = 0
df.Sex2.loc[df.Sex == 'female'] = 1
df.Sex2.loc[df.Sex == 'male'] = 0
df
```

```python
df['Embarked2'] = 0
df.Embarked2.loc[df.Embarked == 'C'] = 0
df.Embarked2.loc[df.Embarked == 'Q'] = 1
df.Embarked2.loc[df.Embarked == 'S'] = 2
df
```

```python
# Ninos
df['Ninos'] = 0
df.Ninos.loc[df.Age < 12] = 1
df.Ninos.loc[df.Age >= 11] = 0
df
```

```python
df.info()
```

```python
#Selección de variables
predictores = df[['Pclass','Age','SibSp','Parch','Fare','Embarked2','Sex2']]
objetivo = df.Survived
```

```python
#Partición para validación cruzada
pred_train, pred_test, res_train, res_test = train_test_split(predictores, objetivo, test_size=0.2)
```

```python
#Modelo SVM
SVM_Model = SVC(gamma='auto', kernel='rbf')
SVM_Model.fit(pred_train, res_train)
```

```python
yhat = SVM_Model.predict(pred_test)
prediccion = list(map(round, yhat))

print('Reales:     ', list(res_test))
print('Prediccion: ', prediccion)
```

```python
true_positive = 0
false_negative = 0
true_negative = 0
false_positive = 0
for predi, real in zip(prediccion, list(res_test)):
  # Verdadero positivo sobrevivió, sobrevivió
  if predi == 1 and real == 1:
    true_positive += 1
  # Falso positivo sobrevivió, murió
  if predi == 1 and real == 0:
    false_positive += 1
  # Verdadero negativo murió, murió
  if predi == 0 and real == 0:
    true_negative += 1
  # Falso negativo murió, sobrevivió
  if predi == 0 and real == 1:
    false_negative += 1

print('Verdadero Positivo: ', true_positive)
print('Falso Positivo: ', false_positive)
print('Verdadero negativo', true_negative)
print('Falso negativo', false_negative)
```

```python
#print (f'Precisión: {SVM_Model.score(pred_test, res_test):.4f}')

precision = true_positive/(true_positive + false_positive)
recall = true_positive/(false_negative + true_positive)

print('Precisión: ', precision)
print('Exhaustividad', recall)
```
